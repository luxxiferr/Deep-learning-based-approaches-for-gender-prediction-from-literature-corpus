#LOADING DATASET

```python
import tensorflow as tf
from tensorflow.keras import layers
import pandas as pd
from datasets import load_dataset
```

```python
path = "/content/drive/MyDrive/AppsI_project/data_path/"

train = load_dataset("csv", data_files= path + "train_nn2.csv" ,sep = ";" )
dev = load_dataset("csv", data_files= path + "dev_nn.csv" ,sep = ";" )
test = load_dataset("csv", data_files= path + "test_nn2.csv" ,sep = ";")

print(train)
print(test)
print(train['train'][0])
print(test['train'][0])
```

#PREPROCESS

```python
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("distilbert-base-uncased")
```

```python
def preprocess_function(examples):
    return tokenizer(examples['text'], truncation=True)
```

```python
tokenized_train = train.map(preprocess_function, batched=True, remove_columns=['Unnamed: 4']) #remove empty features
tokenized_test = test.map(preprocess_function, batched=True)

print(tokenized_train)
print(tokenized_test)
```

```python
from transformers import DataCollatorWithPadding

data_collator = DataCollatorWithPadding(tokenizer=tokenizer)
```

# EVALUATE

```python
import evaluate

accuracy = evaluate.load("accuracy")
```

```python
import numpy as np


def compute_metrics(eval_pred):
    predictions, labels = eval_pred
    predictions = np.argmax(predictions, axis=1)
    return accuracy.compute(predictions=predictions, references=labels)
```

# TRAIN

```python
id2label = {1: "TRUE", 0: "FALSE"}
label2id = {"TRUE": 1, "FALSE": 0}
```

```python
from transformers import AutoModelForSequenceClassification, TrainingArguments, Trainer

model = AutoModelForSequenceClassification.from_pretrained(
    "distilbert-base-uncased", num_labels=2, id2label=id2label, label2id=label2id
)
```

```python
training_args = TrainingArguments(
    output_dir="appsI",
    learning_rate=1e-5,
    per_device_train_batch_size=1,
    per_device_eval_batch_size=188,
    num_train_epochs=3,
    weight_decay=0.0001,
    logging_strategy = "epoch",
    evaluation_strategy = "epoch",
    save_strategy="epoch",
    warmup_steps = 500,
    load_best_model_at_end=True,
    push_to_hub=True,
)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_train['train'],
    eval_dataset=tokenized_test['train'],
    tokenizer=tokenizer,
    data_collator=data_collator,
    compute_metrics=compute_metrics,
)

trainer.train()
```

```python
eval_results = trainer.evaluate()
print(eval_results)
```
